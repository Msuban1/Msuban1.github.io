# Data visualisation

1. Sales by Category Bar Chart	Compare sales across product categories
2. Sales by Region shows which geographic areas are generating the most revenue
3. Variable Relationships	Correlation Heatmap	Show how quantity, price, and total amount relate (same as heatmap in the data exploration)

```python
# Mount Google Drive
from google.colab import drive
import os
drive.mount('/content/drive')

# Set and create save path
save_path = '/content/drive/My Drive/data_viz_outputs'
os.makedirs(save_path, exist_ok=True)

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create DataFrame
data = {
    "Date": pd.to_datetime([
        "2024-01-05", "2024-01-10", "2024-01-12", "2024-02-05", "2024-02-08",
        "2024-02-10", "2024-03-15", "2024-03-18", "2024-03-20", "2024-04-01",
        "2024-04-05", "2024-04-10", "2024-05-01", "2024-05-05", "2024-05-08",
        "2024-05-10", "2024-06-01", "2024-06-05", "2024-06-08", "2024-06-10"
    ]),
    "Product Category": [
        "Electronics", "Electronics", "Electronics", "Electronics", "Books",
        "Electronics", "Clothing", "Clothing", "Electronics", "Books",
        "Electronics", "Electronics", "Clothing", "Electronics", "Electronics",
        "Clothing", "Electronics", "Books", "Electronics", "Electronics"
    ],
    "Region": [
        "North", "South", "West", "East", "North", "North", "South", "West", "East", "North",
        "South", "East", "South", "West", "North", "South", "East", "North", "South", "East"
    ],
    "Quantity": [1,2,1,1,3,1,2,1,1,2,2,1,1,1,1,3,1,4,1,1],
    "Price": [800,600,100,500,20,800,50,25,200,20,600,500,50,100,800,25,200,20,600,500],
    "Total_Amount": [
        800,1200,100,500,60,800,100,25,200,40,1200,500,
        50,100,800,75,200,80,600,500
    ]
}
df = pd.DataFrame(data)

# Sales by Product Category
category_sales = df.groupby("Product Category")["Total_Amount"].sum().sort_values(ascending=False)

plt.figure(figsize=(6,4))
category_sales.plot(kind='bar', color="skyblue")
plt.title("Total Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Sales ($)")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.savefig(f"{save_path}/sales_by_product_category.png")
plt.show()

# Sales by Region
region_sales = df.groupby("Region")["Total_Amount"].sum().sort_values(ascending=False)

plt.figure(figsize=(6,4))
region_sales.plot(kind='bar', color="lightgreen")
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales ($)")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.savefig(f"{save_path}/sales_by_region.png")
plt.show()

# Correlation Heatmap
plt.figure(figsize=(6,5))
sns.heatmap(df[["Quantity", "Price", "Total_Amount"]].corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5, square=True)
plt.title("Correlation Heatmap: Quantity, Price, Total Amount")
plt.tight_layout()
plt.savefig(f"{save_path}/correlation_heatmap.png")
plt.show()
```
